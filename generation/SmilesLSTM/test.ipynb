```python
import plotly.graph_objects as go
import polars as pl
import os


training_dir = "./reports/training/"

metrics_df = pl.read_csv(os.path.join(training_dir, "version_0/metrics.csv"))
train_loss = metrics_df.select("step", "train_loss_step").drop_nulls().cast(pl.Float32)
valid_loss = metrics_df.select("step", "valid_loss").drop_nulls().cast(pl.Float32)

train_step = train_loss[:, 0]
train_loss = train_loss[:, 1]
valid_step = valid_loss[:, 0]
valid_loss = valid_loss[:, 1]

plot = [
    go.Scatter(
        x=train_step, y=train_loss, name="training loss", line=dict(color="#1F77B4")
    ),
    go.Scatter(
        x=valid_step,
        y=valid_loss,
        name="validation loss",
        mode="markers",
        marker=dict(
            size=15,
            color="orange",
            symbol="triangle-up",
            line=dict(width=2, color="black"),
        ),
    ),
]
layout = go.Layout(
    title="Training History",
    width=700,
    height=550,
    font=dict(size=15, family="sans-serif"),
    plot_bgcolor="white",
    xaxis=dict(
        title="# of batches",
        linewidth=1.5,
        mirror=True,
        ticks="inside",
        linecolor="black",
        ticklen=5,
        tickwidth=2,
        tickformat="d",
        minor=dict(tickcolor="black", ticks="inside"),
        range=(0, train_step.max() + train_step.max() * 0.05),
    ),
    yaxis=dict(
        title="Loss",
        linewidth=1.5,
        mirror=True,
        ticks="inside",
        linecolor="black",
        ticklen=5,
        tickwidth=2,
        minor=dict(tickcolor="black", ticks="inside"),
        range=(0, train_loss.max() + train_loss.max() * 0.05),
    ),
    legend=dict(
        x=0.95,
        y=0.95,
        xanchor="right",
        yanchor="top",
        bordercolor="black",
        borderwidth=1,
    ),
)
fig = go.Figure(data=plot, layout=layout)
fig.write_image()

fig.show()
```
